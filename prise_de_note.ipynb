## Merge Sorte 





```python
def merge_sort(numbers: list[int | float]):
    # Simple cases
    if len(numbers) < 2:
        return numbers
    # Divide
    M = len(numbers) // 2
    A = merge_sort(numbers[:M])
    B = merge_sort(numbers[M:])
    # Conquer
    sorted_numbers = []
    while A and B:
        sorted_numbers += [
            A.pop(0) if A[0] < B[0] else B.pop(0)
        ]

    return sorted_numbers + A + B
```

- le but est de prendre une liste : si la liste est < à 2 on return directement la liste car c'est deja trier
- si on divide pour des listes plus grande

```python

#A = [4, 2, 6, 3, 1]

# on cherche d'abord le max et le min puis on trouve autre 

def min_max(A: list[list]) -> tuple[float, float]:
    #simple cases
    # le but de cette partie c'est de voir si la liste == 1 on renvoi des tuples 
    if len(A) == 1:
        return A[0], A[0]
    #Divide
    M = len(A) // 2
    min1, max1 = min_max(A[:M])
    min2, max2 = min_max(A[M:])
    
    #Conquer
    # on peut utiliser les fonctions min et max
    return min(min1, min2), max(max1, max2)
    
 
A, B = min_max([1, 3, 2, 7, 4, 9, 8, 3])
B - A
```

# Quicksort

on choisit un pivot puis on [3 1 6 2 4 3]



```python
def quicksort(A):
    #simples cases
    if len(A) < 2:
        return A
    # divide (with pivot this time!)
    n = len(A) // 2
    pivot = A[n]
    low, high = [], []

    for i in range(1, len(A)):
        if A[i] <= pivot:
            low += [A[i]]
        else: 
            high += [A[i]]
    
    # conquer
    return quicksort(low) + pivot + quicksort(high)

```

# karatsuba's trick for multiplication





```python
def karatsuba(x, y):
    return x * y

karatsuba(34, 13)
```

```python
# solution 
# Complexity: O(n^(log_2 3))
def karatsuba(x, y):
    if x < 10 or y < 10:
        return x * y

    x_str, y_str = str(x), str(y)
    m = max(len(x_str), len(y_str)) // 2
    x_high, x_low = int(x_str[:-m]), int(x_str[-m:])
    y_high, y_low = int(y_str[:-m]), int(y_str[-m:])

    a = karatsuba(x_high, y_high)
    b = karatsuba(x_low, y_low)
    c = karatsuba(x_high + x_low, y_high + y_low) - a - b
    return a * 10 ** (2 * m) + c * 10 ** m + b

karatsuba(34, 13)
```

```python
# solution 

def max_sum(A: list[float]) -> float:
    if len(A) < 2:
        return max(sum(A), 0)

    m = len(A) // 2
    left = max_sum(A[:m])
    right = max_sum(A[m:])

    crossing = A[m]
    for i in [-1, 1]:
        attempt = crossing
        for n in A[m + i::i]:
            attempt += n
            if attempt > crossing:
                crossing = attempt
    return max(crossing, left, right)

max_sum([-2,1,-3,4,-1,2,1,-5,4])
```
